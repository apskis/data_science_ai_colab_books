## What is AI?

- Artificial Intelligence (AI): systems that perform tasks requiring human-like intelligence.

- Examples: speech recognition, image classification, recommendation systems, and natural language processing.

- AI is the umbrella term → includes Machine Learning (ML), Deep Learning, and Generative AI.

## Where AI and ML Intersect

- Machine Learning (ML) is a subset of AI.

- ML focuses on algorithms that learn patterns from data.

- AI = the broader goal; ML = one way to achieve it.

## Generative AI

- Today, when most people say AI, they often mean Generative AI.

- Generative AI creates new content: text, images, audio, or code.

- Examples: ChatGPT, DALL·E, Stable Diffusion.

- Built on deep learning models (like transformers).

## What is an LLM?

- A Large Language Model (LLM) is a type of AI trained to understand and generate human-like text.

- Examples: GPT, Claude, LLaMA.

- They are built on deep learning architectures (especially transformers).

## Training Process

- Data: LLMs are trained on massive text datasets (books, articles, websites).

- Goal: Learn statistical patterns of language.

- Method: Predict the next word in a sequence.

- Example: Given “Data science is …”, the model predicts likely continuations.

## Transformer Architecture (Simplified)

- Attention mechanism: Helps the model focus on the most relevant words in context.

- Layers: Multiple stacked layers refine understanding of meaning and relationships.

- Parameters: Billions of adjustable weights capture patterns in language.

## What LLMs Can Do

- Understand: Classify, summarize, answer questions.

- Generate: Write essays, code, poetry, or dialogue.

- Adapt: Fine-tuned for specific tasks (e.g., medical Q&A, customer support).

## Key Idea

- LLMs don’t “think” like humans.

- They predict text based on patterns learned from data.

- The result often feels intelligent because language itself encodes knowledge.

```python
# Linear Regression Example

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 5, 4, 5])

# fit a model
model = LinearRegression()
model.fit(X, y)

# predictions
y_pred = model.predict(X)

# plot
plt.figure(figsize=(6, 4))
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.xlabel('Dog Weight')
plt.ylabel('Amount of Food Eaten');
```

```python
model.predict([[4.5]])
```

```python
# Generative AI Example

from transformers import pipeline

# load a small language model
generator = pipeline("text-generation", model='distilbert/distilgpt2')
```

```python
# generate text
prompt = "Data science helps businesses because"
generated_output = generator(prompt, max_length=30, num_return_sequences=5)
```

```python
generated_output[0]['generated_text']
```
